1.Find the value of T(2) for the recurrence relation T(n) = 3T(n-1) + 12n, given that T(0)=5.



### Step 1: Calculate \( T(1) \)
Using the recurrence relation for \( n = 1 \):

\[
T(1) = 3T(0) + 12 \times 1
\]
\[
T(1) = 3 \times 5 + 12 = 15 + 12 = 27
\]

### Step 2: Calculate \( T(2) \)
Now, using the recurrence relation for \( n = 2 \):

\[
T(2) = 3T(1) + 12 \times 2
\]
\[
T(2) = 3 \times 27 + 24 = 81 + 24 = 105
\]

### Final Answer:
\[
T(2) = 105
\]

2. Given a recurrence relation, solve it using the substitution method:
a. T(n) = T(n-1) + c
b. T(n) = 2T(n/2) + n
c. T(n) = 2T(n/2) + c
d. T(n) = T(n/2) + c

Let's solve each of the given recurrence relations using the **substitution method**:

### a. \( T(n) = T(n-1) + c \)

We want to find the pattern by unrolling the recurrence:

- Start with \( T(n) = T(n-1) + c \).
- Substitute \( T(n-1) \) using the recurrence:
  \[
  T(n) = (T(n-2) + c) + c = T(n-2) + 2c
  \]
- Substitute \( T(n-2) \):
  \[
  T(n) = T(n-3) + 3c
  \]
- Continuing this process, after \( k \) steps, we get:
  \[
  T(n) = T(n-k) + kc
  \]
- Eventually, when \( k = n \), \( T(0) \) will be the base case. Assuming \( T(0) = T_0 \), we have:
  \[
  T(n) = T_0 + nc
  \]

Thus, the solution to the recurrence is:
\[
T(n) = T_0 + nc
\]

### b. \( T(n) = 2T(n/2) + n \)

We can solve this recurrence by unrolling:

- Start with \( T(n) = 2T(n/2) + n \).
- Substitute \( T(n/2) \) using the recurrence:
  \[
  T(n) = 2(2T(n/4) + n/2) + n = 4T(n/4) + n + n = 4T(n/4) + 2n
  \]
- Substitute \( T(n/4) \):
  \[
  T(n) = 4(2T(n/8) + n/4) + 2n = 8T(n/8) + n + 2n = 8T(n/8) + 3n
  \]
- Continuing this process, after \( k \) steps, we get:
  \[
  T(n) = 2^k T(n / 2^k) + kn
  \]
- Eventually, \( n / 2^k = 1 \), which happens when \( k = \log_2 n \). At this point, we reach the base case \( T(1) \), which is a constant. Let \( T(1) = T_1 \), so:
  \[
  T(n) = 2^{\log_2 n} T(1) + n \log_2 n
  \]
  Simplifying \( 2^{\log_2 n} = n \), we get:
  \[
  T(n) = n T_1 + n \log_2 n
  \]
  
Thus, the solution is:
\[
T(n) = O(n \log n)
\]

### c. \( T(n) = 2T(n/2) + c \)

This recurrence is similar to part (b), but with a constant \( c \) instead of \( n \). Let's unroll it:

- Start with \( T(n) = 2T(n/2) + c \).
- Substitute \( T(n/2) \):
  \[
  T(n) = 2(2T(n/4) + c) + c = 4T(n/4) + 2c + c = 4T(n/4) + 3c
  \]
- Substitute \( T(n/4) \):
  \[
  T(n) = 4(2T(n/8) + c) + 3c = 8T(n/8) + 4c + 3c = 8T(n/8) + 7c
  \]
- Continuing this process, after \( k \) steps, we get:
  \[
  T(n) = 2^k T(n / 2^k) + (2^k - 1)c
  \]
- Eventually, when \( k = \log_2 n \), \( n / 2^k = 1 \), so we get:
  \[
  T(n) = n T(1) + (n - 1)c
  \]
  Assuming the base case is constant, say \( T(1) = T_1 \), the solution is:
  \[
  T(n) = O(n)
  \]

### d. \( T(n) = T(n/2) + c \)

Let's solve this recurrence using substitution:

- Start with \( T(n) = T(n/2) + c \).
- Substitute \( T(n/2) \):
  \[
  T(n) = (T(n/4) + c) + c = T(n/4) + 2c
  \]
- Substitute \( T(n/4) \):
  \[
  T(n) = T(n/8) + 3c
  \]
- Continuing this process, after \( k \) steps, we get:
  \[
  T(n) = T(n / 2^k) + kc
  \]
- Eventually, when \( k = \log_2 n \), \( n / 2^k = 1 \), so we get:
  \[
  T(n) = T(1) + c \log_2 n
  \]
  Assuming \( T(1) = T_1 \), the solution is:
  \[
  T(n) = O(\log n)
  \]

### Summary of Solutions:

1. \( T(n) = T(n-1) + c \) → \( T(n) = T_0 + nc \)
2. \( T(n) = 2T(n/2) + n \) → \( T(n) = O(n \log n) \)
3. \( T(n) = 2T(n/2) + c \) → \( T(n) = O(n) \)
4. \( T(n) = T(n/2) + c \) → \( T(n) = O(\log n) \)

3. Given a recurrence relation, solve it using the recursive tree approach:
a. T(n) = 2T(n-1) +1
b. T(n) = 2T(n/2) + n

### Solving Recurrence Relations Using the Recursive Tree Approach

Let's break down both recurrence relations using the **recursive tree approach**, which helps visualize the work done at each level of recursion and the total cost.

### a. \( T(n) = 2T(n-1) + 1 \)

#### Step 1: Unroll the recurrence

To understand how the function behaves at each step, let's unroll the recurrence:

- \( T(n) = 2T(n-1) + 1 \)
- \( T(n-1) = 2T(n-2) + 1 \)
- \( T(n-2) = 2T(n-3) + 1 \)
- ...
- \( T(1) = 2T(0) + 1 \)

#### Step 2: Visualizing the recursive tree

At the top of the tree (level 0), we have \( T(n) = 2T(n-1) + 1 \). The work done at this level is 1 (constant). Then, at level 1, we have two calls, each contributing \( T(n-1) \), and each of those performs 1 unit of work. At level 2, we have four calls to \( T(n-2) \), each doing 1 unit of work. The process continues down until we reach the base case \( T(0) \), which is assumed to be a constant.

#### Step 3: Tree structure

- At level 0: 1 call to \( T(n) \), performing 1 unit of work.
- At level 1: 2 calls to \( T(n-1) \), each performing 1 unit of work.
- At level 2: 4 calls to \( T(n-2) \), each performing 1 unit of work.
- At level 3: 8 calls to \( T(n-3) \), each performing 1 unit of work.
- ...
- At level \( k \): \( 2^k \) calls to \( T(n-k) \), each performing 1 unit of work.

#### Step 4: Total work

The total work is the sum of the work done at each level. The number of levels is \( n \) (since the argument decreases by 1 each time). The work done at each level is \( 2^k \) (where \( k \) is the level number), and the total work is:

\[
W(n) = 1 + 2 + 4 + 8 + \dots + 2^{n-1}
\]

This is a geometric series, and its sum is:

\[
W(n) = 2^n - 1
\]

Thus, the total work done is \( O(2^n) \).

#### Final Answer:
\[
T(n) = O(2^n)
\]

### b. \( T(n) = 2T(n/2) + n \)

#### Step 1: Unroll the recurrence

Let's unroll the recurrence to understand the work done at each level:

- \( T(n) = 2T(n/2) + n \)
- \( T(n/2) = 2T(n/4) + n/2 \)
- \( T(n/4) = 2T(n/8) + n/4 \)
- ...

#### Step 2: Visualizing the recursive tree

At the top level (level 0), we have \( T(n) = 2T(n/2) + n \), and it performs \( n \) work. Then, at level 1, we have two calls to \( T(n/2) \), each performing \( n/2 \) work, for a total of \( n \) work at that level. At level 2, we have four calls to \( T(n/4) \), each performing \( n/4 \) work, again totaling \( n \) work at that level. The process continues down until we reach the base case.

#### Step 3: Tree structure

- At level 0: 1 call to \( T(n) \), performing \( n \) units of work.
- At level 1: 2 calls to \( T(n/2) \), each performing \( n/2 \) units of work, so total work at this level is \( n \).
- At level 2: 4 calls to \( T(n/4) \), each performing \( n/4 \) units of work, so total work at this level is \( n \).
- At level 3: 8 calls to \( T(n/8) \), each performing \( n/8 \) units of work, so total work at this level is \( n \).
- ...
- At level \( k \): \( 2^k \) calls to \( T(n / 2^k) \), each performing \( n / 2^k \) work, so total work at this level is \( n \).

#### Step 4: Total work

The total work at each level is \( n \), and there are \( \log_2 n \) levels (because at each level, the problem size halves). So the total work is:

\[
W(n) = n \times \log_2 n
\]

Thus, the total work done is \( O(n \log n) \).

#### Final Answer:
\[
T(n) = O(n \log n)
\]

### Summary of Results:

1. \( T(n) = 2T(n-1) + 1 \) → \( T(n) = O(2^n) \)
2. \( T(n) = 2T(n/2) + n \) → \( T(n) = O(n \log n) \)
